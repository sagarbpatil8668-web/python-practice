```python
# STRING INDEXING IN PYTHON
# ------------------------
# Indexing means accessing individual characters from a string
# Index starts from 0 (positive indexing)
# Negative indexing starts from -1 (from right to left)

s = "PYTHON"

# --------------------------------------------------
# Positive Indexing
# P  Y  T  H  O  N
# 0  1  2  3  4  5

print(s[0])   # P
print(s[1])   # Y
print(s[2])   # T
print(s[5])   # N

# --------------------------------------------------
# Negative Indexing
# P   Y   T   H   O   N
# -6 -5  -4  -3  -2  -1

print(s[-1])  # N
print(s[-2])  # O
print(s[-6])  # P

# --------------------------------------------------
# Using len() with Positive Index
# Last index = len(string) - 1

print(len(s))         # 6
print(s[len(s) - 1])  # N

# --------------------------------------------------
# Using len() for Negative Index Understanding
# First character index = -len(string)

print(s[-len(s)])     # P

# --------------------------------------------------
# Looping with Index

for i in range(len(s)):
    print("Index", i, ":", s[i])

# --------------------------------------------------
# IndexError Example

# print(s[6])    # ❌ IndexError: string index out of range
# print(s[-7])   # ❌ IndexError

# --------------------------------------------------
# Indexing with User Input

text = input("Enter a word: ")

print("First character:", text[0])
print("Last character :", text[-1])
```

slicing

```python
# STRING SLICING IN PYTHON
# -----------------------
# Slicing is used to extract a part of a string
# Syntax:
# string[start : stop : step]
#
# start → index to begin (inclusive)
# stop  → index to end (exclusive)
# step  → jump value (optional)

s = "PYTHON"

# Index reference
# P  Y  T  H  O  N
# 0  1  2  3  4  5
#-6 -5 -4 -3 -2 -1

# --------------------------------------------------
# 1. Basic Slicing (start : stop)

print(s[0:4])    # PYTH
print(s[1:5])    # YTHO
print(s[2:6])    # THON

# stop index is NOT included

# --------------------------------------------------
# 2. Slicing with start only (start : )

print(s[2:])     # THON
print(s[4:])     # ON

# --------------------------------------------------
# 3. Slicing with stop only ( : stop)

print(s[:3])     # PYT
print(s[:5])     # PYTHO

# --------------------------------------------------
# 4. Full string slice

print(s[:])      # PYTHON

# --------------------------------------------------
# 5. Slicing with step (start : stop : step)

print(s[0:6:1])  # PYTHON
print(s[0:6:2])  # PTO
print(s[1:6:2])  # YHN

# --------------------------------------------------
# 6. Negative slicing

print(s[-6:-1])  # PYTHO
print(s[-5:-2])  # YTH
print(s[-4:])    # THON

# --------------------------------------------------
# 7. Reverse a string using slicing

print(s[::-1])   # NOHTYP

# --------------------------------------------------
# 8. Partial reverse

print(s[4:1:-1]) # OHT
print(s[-1:-7:-1]) # NOHTYP

# --------------------------------------------------
# 9. Slicing with input

text = input("Enter a string: ")
print("First 3 chars :", text[:3])
print("Last 3 chars  :", text[-3:])
print("Reversed text :", text[::-1])
```

IMP in strings 

+ve step → start < stop → defaults (0 → len) 

-ve step → start > stop → defaults (len-1 → None)

Same start & stop → empty string

Wrong direction → empty string

Step = 0 → ERROR

Never use -1 as stop when step is negative
Leave stop empty (:) to go fully backward.

The stop index is ALWAYS EXCLUDED

This rule does not change for positive or negative step.
